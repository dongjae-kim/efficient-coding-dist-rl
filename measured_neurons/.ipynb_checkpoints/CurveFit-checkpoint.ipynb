```python
# load
import scipy.io as sio
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats.distributions import  t
import matplotlib.pyplot as plt


def func(x,a,b,c):
    return b / (1 + np.exp(-(x-c)*a))


juiceAmounts = [0.1, 0.3, 1.2, 2.5, 5, 10, 20];
dat = sio.loadmat('dat_eachneuron.mat')
dat = dat['dat']

ps = []
ps_ucb = []
ps_lcb = []

ps_total = []

for i in range(40):
    dats = dat[0,i][0]
    datas_x = []
    datas_y = []

    for count, j in enumerate(juiceAmounts):
        y_to_extend = np.squeeze(dats[np.where(~np.isnan(dats[:,count])),count])
        x_to_extend = (np.ones(y_to_extend.shape)*j).tolist()    
        datas_x.extend(x_to_extend)
        datas_y.extend(y_to_extend)
    
    x =datas_x
    y = datas_y


    initial_guess = [1, 1, 0]
    pars, pcov = curve_fit(func, x, y, p0=initial_guess, bounds = (0,np.inf))

#     alpha = 0.05 # 95% confidence interval = 100*(1-alpha)
    alpha = 0.1 # 90% confidence interval = 100*(1-alpha)

    n = len(y)    # number of data points
    p = len(pars) # number of parameters

    dof = max(0, n - p) # number of degrees of freedom

    # student-t value for the dof and confidence level
    tval = t.ppf(1.0-alpha/2., dof) 
    ps_ =[]
    ps_ucb_ =[]
    ps_lcb_ =[]
    for i, p,var in zip(range(n), pars, np.diag(pcov)):
        sigma = var**0.5
        ps_.append(p)
        ps_ucb_.append( p + sigma*tval)
        ps_lcb_.append( p - sigma*tval)
        
    
    ps.append(ps_)
    ps_ucb.append(ps_ucb_)
    ps_lcb.append(ps_lcb_)
    
    ps_total_ = []
    tts = np.linspace(-tval, tval, 121)
    for ttval in tts:
        ps_total_t = []
        #  40 neurons student t-test value linspace
        for i, p,var in zip(range(n), pars, np.diag(pcov)):
                sigma = var**0.5
                ps_total_t.append(p + sigma*ttval)

#                 if np.abs(sigma*ttval) < 0.0001:
#                     print(sigma*ttval)

        ps_total_.append(ps_total_t)
    
    ps_total.append(ps_total_)
    
    

```

```python
alpha = 0.1 # 95% confidence interval = 100*(1-alpha)
tval = t.ppf(1.0-alpha/2., dof) 
tval
```

```python
sio.savemat('curve_fit_parameters_new.mat', {'ps':ps,'ps_ucb':ps_ucb,'ps_lcb':ps_lcb, 'ps_total':ps_total}) # post fix '_new' to prevent overwriting
```
